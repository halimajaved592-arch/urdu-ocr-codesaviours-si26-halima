```python
1. What is OCR (Optical Character Recognition)?

Optical Character Recognition (OCR) is a technology that converts text
 from images, scanned documents, or photographs into editable and searchable digital text.
  It uses image processing and artificial intelligence to recognize letters, words, and numbers automatically.
   OCR saves time by eliminating the need to manually type printed or handwritten text.
```

```python
2. Why is Urdu OCR harder than English OCR?

Urdu OCR is more challenging because Urdu is written from right to left and
 many letters change their shape depending on their position in a word.
  Several Urdu characters look very similar and differ only by small dots, making them difficult for OCR systems
   to distinguish. In addition, variations in fonts, handwriting styles, and image quality
    make accurate recognition more difficult than in English.
```

```python
. What are two real-world situations where Urdu OCR would be useful?

Urdu OCR can be used to digitize old books, newspapers, and historical documents
 so they can be searched and preserved electronically. It is also useful for converting
  printed forms, receipts, and government documents into editable digital text, reducing manual data
  entry and improving efficiency. These applications help organizations save time and
   make information easier to access.
```

```python
import os
os.makedirs("data/realgt", exist_ok=True)
```

```python
import os
import csv

image_folder = "data/raw/other"
text_folder = "data/realgt"

labels = []

for image_name in sorted(os.listdir(image_folder), key=lambda x: int(os.path.splitext(x)[0])):
    if image_name.endswith(".png"):
        image_number = os.path.splitext(image_name)[0]
        txt_file = os.path.join(text_folder, image_number + ".txt")

        if os.path.exists(txt_file):
            with open(txt_file, "r", encoding="utf-8") as f:
                text = f.read().strip()

            labels.append({
                "image": f"data/raw/other/{image_name}",
                "text": text
            })

with open("data/labels.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["image", "text"])
    writer.writeheader()
    writer.writerows(labels)

print(f"labels.csv created with {len(labels)} entries!")
```
